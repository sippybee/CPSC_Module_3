```python
class Customer:
    def __init__(self, name, balance=0.0):
        self.name = name
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount
        return self.balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            raise RuntimeError('Amount greater than available balance')
        self.balance -= amount
        return self.balance
```

```python
j = Customer("John", 1000)
```

```python
j.balance
```

```python
j.deposit(100)
```

```python
j.withdraw(900)
```

```python
dir(j)
```

```python
j.balance
```

```python
class RoomArea:
    def __init__ (self, width, length):
        self.width = width
        self.length = length
        self.area = self.area()
 
    def area (self):
        self.area = self.length * self.width
        return self.area
```

```python
r1 = RoomArea(5,6)
```

```python
r1.area
```
