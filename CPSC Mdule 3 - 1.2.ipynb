```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com/"
```

```python
response = requests.get(url)
bsobj = BeautifulSoup(response.content, "html.parser")
```

```python
number_of_pages_raw =  bsobj.find('li', class_='current').get_text().split()[-1]
number_of_pages = int(number_of_pages_raw)
number_of_pages
```

```python
def get_book_info_link(book):
    try:
        title = book.find('h3').text
    except:
        title = "No title"
    try:
        price_raw = book.find('p', class_='price_color').get_text()
        price = float(price_raw.strip('£'))
    except:
        price = 0.00
    try:
        link = book.find("a")["href"]
    except:
        link = "No link"
    return title, price, link
```
