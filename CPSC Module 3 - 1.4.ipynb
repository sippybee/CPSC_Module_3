```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import chromedriver_binary
import chromedriver_autoinstaller
import time
```

```python
def get_details(card):
    print("----->",card)
    try:
        title = card.find_element(By.TAG_NAME, "h2").text
    except:
        title = "No title"
    try:
        href = card.find_element(By.TAG_NAME, "a").get_attribute("href")
        asin_raw = href.split("dp")[2].split("%")[1]
        if len(asin_raw) == 12:
            asin = asin_raw
    except:
        asin ="No asin"
    try:
        price_raw = card.find_element(By.CLASS_NAME,"a-price").text.split()
        price = f"{price_raw[0]}.{price_raw[1]}"
    except:
        price = "No price"
    if title != "No title" and asin !="No asin":
        return [title, asin, price]
    else:
        pass
```

```python
chromedriver_autoinstaller.install()
driver = Chrome()
driver.get("https://www.amazon.com")
wait = WebDriverWait(driver, 10)
search_box = wait.until(EC.presence_of_element_located((By.ID, "twotabsearchtextbox")))
search_box.send_keys("headphones")
search_box.send_keys(Keys.ENTER)
cards = driver.find_elements(By.CLASS_NAME, "s-card-container")
for card in cards:
    try:
        title = card.find_element(By.TAG_NAME, "h2").text
    except:
        title = "No title"
    try:
        href = card.find_element(By.TAG_NAME, "a").get_attribute("href")
        asin_raw = href.split("dp")[2].split("%")[1]
        if len(asin_raw) == 12:
            asin = asin_raw
    except:
        asin ="No asin"
    try:
        price_raw = card.find_element(By.CLASS_NAME,"a-price").text.split()
        price = f"{price_raw[0]}.{price_raw[1]}"
    except:
        price = "No price"
    if title != "No title" and asin !="No asin":
        print([title, asin, price])
time.sleep(3)
driver.find_element(By.CLASS_NAME, "s-pagination-next").click()
time.sleep(3)
driver.find_element(By.CLASS_NAME, "s-pagination-next").click()
```
