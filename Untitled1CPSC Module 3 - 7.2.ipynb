```python
from recbole.quick_start import run_recbole
from recbole.config import Config
from recbole.data import create_dataset
import ast
import os
import pandas as pd
```

```python
def build_atomic_file(csv_path: str, dataset_name: str= "PP_users", data_root: str ="./dataset"):
    df = pd.read_csv(csv_path)
    # The 'items' and 'ratings' columns are stored as stringified Python lists
    df["items"] = df["items"].apply(ast.literal_eval)
    df["ratings"] = df["ratings"].apply(ast.literal_eval)
    # Explode into long format: one row per (user, item, rating)
    long_df = df[["u", "items", "ratings"]].explode(["items", "ratings"])
    long_df.columns = ["user_id", "item_id", "rating"]
    long_df["item_id"] = long_df["item_id"].astype(int)
    long_df["rating"] = long_df["rating"].astype(float)
     # Build the folder RecBole expects: dataset/PP_users/PP_users.inter
    out_dir = os.path.join(data_root, dataset_name)
    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, f"{dataset_name}.inter")
     # RecBole atomic files use "field:type" column headers, tab-separated
    atomic = long_df.rename(columns={
        "user_id": "user_id:token",
        "item_id": "item_id:token",
        "rating": "rating:float",
    })
    atomic.to_csv(out_path, sep="\t", index=False)
 
    print(f"Wrote {len(atomic):,} interactions to {out_path}")
    return data_root, dataset_name
 
```

```python
def run_model(data_root: str, dataset_name: str):
    config_dict = {
        "data_path": data_root + "/",
        "USER_ID_FIELD": "user_id",
        "ITEM_ID_FIELD": "item_id",
        "RATING_FIELD": "rating",
        "load_col": {"inter": ["user_id", "item_id", "rating"]},
        "epochs": 20,
        "train_batch_size": 4096,
        "eval_args": {"split": {"RS": [0.8, 0.1, 0.1]}, "group_by": "user"},
        "metrics": ["Recall", "NDCG", "Precision", "Hit"],
        "topk": [5, 10],
        # RecBole needs ONE metric to decide "which epoch is best" during training.
        # This must be one of the metrics listed above -- it defaults to 'MRR@10',
        # which causes a KeyError if MRR isn't in the metrics list.
        "valid_metric": "NDCG@10",
    }
 
    result = run_recbole(model="BPR", dataset=dataset_name, config_dict=config_dict)
    print(result)
    return result
```

```python
if __name__ == "__main__":
    CSV_PATH = "PP_users.csv"  # update this path if the file lives elsewhere
 
    data_root, dataset_name = build_atomic_file(CSV_PATH)
    run_model(data_root, dataset_name)
```
