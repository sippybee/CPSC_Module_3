```python
with open("article.txt",'r', encoding='utf-8') as f:
    text = f.read()

new_documents = sent_tokenize(text)

vectorizer = CountVectorizer(stop_words="english")
doc_term_matrix = vectorizer.fit_transform(new_documents)

num_topics = 3
new_lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
new_lda.fit(doc_term_matrix)

words = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(new_lda.components_):
    top_words_idx = topic.argsort()[:-5:-1]
    top_words = [words[i] for i in top_words_idx]
    print(f"Topic #{topic_idx+1}: {', '.join(top_words)}")
```
