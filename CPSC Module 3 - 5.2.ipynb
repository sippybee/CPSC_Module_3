```python
def unique_ratio(paragraph):
    tokens = [t.lower() for t in word_tokenize(paragraph) if t.isalpha()]
    if len(tokens) == 0:
        return 0
    return len(set(tokens))/len(tokens)

para_ratios = [ (p, unique_ratio(p)) for p in paragraphs]
max(para_ratios, key=lambda x : x[1])
```

```python
# tf = times word appears in the doc / total words in doc
# idf = log(total number of docs / documents containing the word)
# tf-idf = tf * idf
# the - all docs
# Obesity - rare in all docs
```

```python
from nltk.text import TextCollection
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
paragraphs = [p.strip() for p in text.split("\n") if len( p.strip()) > 100]
len(paragraphs)

vectorizer = TfidfVectorizer(stop_words="english", max_features=500)
tfidf_matrix = vectorizer.fit_transform(paragraphs)
feature_names = vectorizer.get_feature_names_out()
```

```python
for i in range(5):
    row = tfidf_matrix[i].toarray().flatten()
    top_indices = row.argsort()[-5:][::-1]
    top_words = [(feature_names[idx], round(row[idx], 3)) for idx in top_indices]
    print(f"\n Paragraph {i+1}: {paragraphs[i][:60]}...")
    print(f"Top Keywords: {top_words}")
```
