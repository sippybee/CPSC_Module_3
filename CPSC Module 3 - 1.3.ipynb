```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
file = open("book_to_scrape.csv")
links = [i.split(",")[-1].strip("\n") for i in file][1:]
```

```python
def get_info(link):
    url = f"https://books.toscrape.com/{link}"
    response = requests.get(url)
    bsobj = BeautifulSoup(response.content, "html.parser")
    try:
        title = bsobj.find("h1").text
    except:
        title = "No title"
    try:
        price_raw = bsobj.find('p', class_='price_color').get_text(strip=True)
        price = float(price_raw.strip("£"))
    except:
        price = 0.00
    try:
        availability_raw = bsobj.find('p', class_='instock availability').get_text(strip=True)
        availability = availability_raw.split("(")[-1].split()[0]
    except:
        availability = 0
    description = bsobj.find("article",{"class":"product_page"}).text.split("Product Description")[1]
    print(description)
    return title, price, availability
```

```python
data = [get_info(i) for i in links[:5]]
```

```python
df = pd.DataFrame(data, columns=["Title","Price","Availability"])
df
```
